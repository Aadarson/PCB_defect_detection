### 🔥 Start GPU Training (50 Epochs)
This will automatically discover `dataset/data.yaml` from what you just unzipped.

```python
from ultralytics import YOLO

# Load Nano model
model = YOLO("yolov8n.pt")

# Train using the massive T4 GPU power
results = model.train(
    data="dataset/data.yaml",  
    epochs=50,
    imgsz=640,
    batch=16,          
    device=0,          
    workers=8,         
    optimizer="AdamW",
    lr0=0.001,
    hsv_h=0.015,       
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10.0,
    translate=0.1,
    scale=0.5,
    mosaic=1.0
)
```
